```python
# Import required libraries
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

Firstly, create and show image with half size in both axis with original ones.

```python
# Open the image file
img = io.imread('00_06.jpg')

# Scale the image to half its size, subtracting every second pixel from each row
little_img = img[::, ::2]
# Then subtracting every second pixel from each column
little_img = little_img[::2, ::]
```

```python
# Display the original image
fig, axs = plt.subplots(1,1, figsize=(3, 3))
axs.imshow(img)
axs.set_title('Original')
plt.show()
```

```python
# Display the little image
fig, axs = plt.subplots(1,1, figsize=(1.5, 1.5))
axs.imshow(little_img)
axs.set_title('Little image')
plt.show()
```

Now, create and show image with double size in both axis with original ones.

```python
# Get the size of the image
height, width, _ = img.shape

# Create an empty array with the shape of the doubled image
big_img = np.zeros((height*2, width*2, 3), dtype=np.uint8)

# Fill the new array with the pixel values from the original image
for i in range(height):
    for j in range(width):
        # We add additionally pixel to original image row and column
        big_img[i*2:i*2+2, j*2:j*2+2] = img[i:i+1, j:j+1,:3]
```

```python
# Display the original image
fig, axs = plt.subplots(1,1, figsize=(3, 3))
axs.imshow(img)
axs.set_title('Original')
plt.show()
```

```python
# Display the big image
fig, axs = plt.subplots(1,1, figsize=(6, 6))
axs.imshow(big_img)
axs.set_title('Big image')
plt.show()
```
